```python
import pandas as pd
df = pd.read_csv("/content/Eye_movement_dataset.csv", sep=";", encoding_errors='ignore')
pd.set_option('display.max_columns', None)
df
```

```python
def min_max(column):
  range_column = [min(column), max(column)]
  return range_column
```

```python
def labels(df, topic):
  return list(dict(df.groupby(topic).groups).keys())
```

```python
metadata = []
types = list(dict(df.dtypes).items())
for item in types:
  if item[1] == "object":
    metadata_column = labels(df, item[0])
  else:
    metadata_column = min_max(df[item[0]])
  metadata.append(metadata_column)
```
